```python
picture=cv2.imread(r"C:\Users\ACER\Desktop\CAT-PICTURE.jpg")
```

# plotting the image of cat

```python
plt.title("cat image")
plt.imshow(picture)
plt.show()
```

# Gray image

```python
gray_picture=cv2.cvtColor(picture,cv2.COLOR_BGR2GRAY)
plt.title("Cat Gray image")
plt.imshow(gray_picture)
plt.show()
```

# inverted image

```python
inverted_image= 255-gray_picture
plt.title("Inverted Image")
plt.imshow(inverted_image)
plt.show()
```

# converting image to blur image 

```python
blurred= cv2.GaussianBlur(inverted_image,(21,21),0)
```

```python
plt.imshow(blurred)
plt.axis(False)
plt.show()
```

# inverted the blur image and converting it into pencil sketch

```python
inverted_blurred=255-blurred
pencil_sketch=cv2.divide(gray_picture,inverted_blurred,scale=256.0)
plt.title("pencil sketch image of cat")
plt.imshow(pencil_sketch)
plt.show()
```

```python
plt.figure(figsize=(14,8))
plt.subplot(1,2,1)
plt.title("original image ", size=25)
plt.imshow(picture)
plt.axis('off')
plt.subplot(1,2,2)
plt.title('sketch',size=25)
rgb_sketch=cv2.divide(gray_picture,inverted_blurred,scale=256.0)
plt.imshow(rgb_sketch)
plt.axis('off')
plt.show()
```
